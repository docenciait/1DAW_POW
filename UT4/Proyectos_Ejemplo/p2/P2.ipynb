# Laboratorio: Análisis de Datos de Ventas con Pandas, NumPy y Matplotlib

## Objetivo:

- Analizar datos de ventas para identificar tendencias, patrones y oportunidades de mejora.
- Visualizar datos para una mejor comprensión y toma de decisiones.

## Requisitos:

- Python 3 instalado
- Bibliotecas Pandas, NumPy y Matplotlib instaladas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Carga los datos
data = pd.read_csv("datos_ventas_extendido.csv")

# Explora los datos
print(data.head())
print(data.info())
print(data.describe())

# Limpia y preprocesa los datos
data["Fecha"] = pd.to_datetime(data["Fecha"])
data["Total"] = data["Cantidad"] * data["Precio unitario"]

# Analiza los datos
ventas_por_mes = data.groupby(pd.Grouper(key="Fecha", freq="M"))["Total"].sum()
print(ventas_por_mes)

productos_mas_vendidos = data.groupby("Producto")["Cantidad"].sum().nlargest(10)
print(productos_mas_vendidos)

promedio_ventas_por_cliente = data.groupby("Cliente")["Total"].mean()
print(promedio_ventas_por_cliente)

# Análisis extendido
ventas_por_region = data.groupby("Region")["Total"].sum()
print(ventas_por_region)

ventas_por_canal = data.groupby("Canal de venta")["Total"].sum()
print(ventas_por_canal)

# Visualiza los datos
plt.figure(figsize=(10, 6))
plt.plot(ventas_por_mes.index, ventas_por_mes.values)
plt.xlabel("Mes")
plt.ylabel("Ventas totales")
plt.title("Ventas totales por mes")
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
plt.bar(productos_mas_vendidos.index, productos_mas_vendidos.values)
plt.xticks(rotation=45, ha="right")
plt.xlabel("Producto")
plt.ylabel("Cantidad vendida")
plt.title("Productos más vendidos")
plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 6))
plt.pie(ventas_por_region, labels=ventas_por_region.index, autopct="%1.1f%%")
plt.title("Ventas por región")
plt.show()

plt.figure(figsize=(8, 6))
plt.pie(ventas_por_canal, labels=ventas_por_canal.index, autopct="%1.1f%%")
plt.title("Ventas por canal de venta")
plt.show()
```

## Mejoras y extensiones:

- Análisis por región y canal de venta: Se han agregado análisis y visualizaciones para las ventas por región y canal de venta. Esto permite identificar qué regiones y canales generan más ingresos.

- Gráficos de tarta: Se han incluido gráficos de tarta para visualizar la distribución de las ventas por región y canal de venta de manera más clara.

- Mayor cantidad de datos: El archivo CSV de ejemplo ahora contiene más datos, lo que permite un análisis más completo y representativo.
