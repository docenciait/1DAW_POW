```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

from pathlib import Path

%matplotlib inline
```

```python
data_path = Path('./data')

data = pd.read_csv(data_path / 'house.csv')
data.head(10)
```

```python
data.groupby('bedrooms').max()
```

```python
np.sort(data['bedrooms'].unique())
```

```python
data.groupby('bedrooms')['sqft_above'].mean().plot(figsize = (20,8), c = 'blue', legend = True)
f = data.groupby('bedrooms')['sqft_living'].mean().plot(figsize = (20,8), c = 'green', legend = True)

f.set_title("Cantidades medias por nº de cuartos de baño")
plt.xticks(np.arange(0, len(data['bedrooms'].unique())), np.sort(data['bedrooms'].unique()));
```
